```python
def clean(df):
    unwanted_cols = ["Province/State", "Lat", "Long"]
    df = df.fillna("")  # handling null cases

    # collapsing provinces
    provinced_locations = df[df.duplicated(["Country/Region"])]
    df = df.drop_duplicates(["Country/Region"])
    for country in provinced_locations["Country/Region"].unique():
        current_country_data = (
            provinced_locations[
                (provinced_locations["Country/Region"] == country)
            ]
            .sum(axis=0)
        )
        current_country_data["Country/Region"] = country
        df = df[df["Country/Region"] != country]
        df = df.append(current_country_data, ignore_index=True)

    df.drop(unwanted_cols, axis=1, inplace=True)
    df = df.T  # transpose

    # columns name as country names
    df.columns = df.iloc[0]
    df = df.drop("Country/Region")

    # The `Date`
    df.reset_index(level=0, inplace=True)
    df.columns.values[0] = "Date"

    # The `Day`
    df.reset_index(level=0, inplace=True)
    df.columns.values[0] = "Day"

    # date type
    df["Date"] = pd.to_datetime(df["Date"])
    return df
```

```python
death_data = pd.read_csv("data/deaths.csv")
death_data = clean(death_data)
```

```python
recovered_data = pd.read_csv("data/recovered.csv")
recovered_data = clean(recovered_data)
```

```python
def gaussian(x, a, b, c):
    from numpy import e
    return a*e**(-(x-b)**2/c)
```

# Fitting

```python
from scipy.optimize import curve_fit
popt, error = curve_fit(gaussian, austria_peak["Day"], austria_peak["Austria"], [8000, 70, 1])
gaussian(60, *popt)
```

```python
fit_x = np.linspace(30, 100, 70)
fit_y = gaussian(fit_x, *popt)
x = austria_peak["Day"]
y = austria_peak["Austria"]
plt.plot(fit_x, fit_y, label="gaussian fit")
plt.plot(x, y, label="data")
plt.xlabel("Dates")
plt.ylabel("Active cases")
plt.legend()
plt.grid()
plt.show()
```
